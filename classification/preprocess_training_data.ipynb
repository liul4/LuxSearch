```python
import pandas as pd
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
df = pd.read_csv('training.1600000.processed.noemoticon.csv', encoding='ISO-8859-1', names=['target','ids','date','flag','user','text'])
df.head()
```

```python
# Remove URLs
df['text'] = df['text'].str.replace(r'http\S+', '')
# Remove @
df['text'] = df['text'].str.replace(r'@\S+', '')
# Remove hashtags
df['text'] = df['text'].str.replace(r'#\S+', '')
# Remove usernames
df['text'] = df['text'].str.replace(r'\@\w+', '')
# Remove numbers
df['text'] = df['text'].str.replace(r'\d+', '') 
# Remove punctuation
df['text'] = df['text'].str.replace(r'[^\w\s]', '') 
# Convert to lowercase
df['text'] = df['text'].str.lower()
#remove storpwords
stop_words = stopwords.words('english')
df['cleaned_text'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
df.head()
```

```python
df.to_csv("cleaned_training_data.csv", encoding="utf-8", index=False)
```
